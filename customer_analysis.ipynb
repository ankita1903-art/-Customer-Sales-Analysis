```python
import pandas as pd
df=pd.read_csv('customer_churn.csv')
total_customers = df['CustomerID'].nunique()

total_revenue = df["TotalCharges"].sum()
sales_by_product = df.groupby("MonthlyCharges")["TotalCharges"].sum()
best_product = (
    df.groupby('Contract')['TotalCharges']
      .sum()
      .sort_values(ascending=False)
      .reset_index(name='total_revenue')
)

top_customer = (
    df.groupby(['CustomerID'])['TotalCharges']
      .sum()
      .sort_values(ascending=False)
      .reset_index()
      .iloc[0]
)

print("\nCUSTOMER SALES ANALYSIS REPORT")
print(f"Total Customers: {total_customers:,}")
print(f"Top Customer: {top_customer['CustomerID']} - ${top_customer['TotalCharges']:,.0f}")
print("TOTAL REVENUE",total_revenue)
print("SALES BY PRODUCT",sales_by_product)
print("BEST PRODUCT",best_product)
```

```python
pivot_table = pd.pivot_table(
    df,
    values='TotalCharges',
    index='Contract',
    columns='PaymentMethod',
    aggfunc='sum',
    fill_value=0
)

print("\nREVENUE PIVOT TABLE")
print(pivot_table)
```

```python
cross_selling = (
    df.groupby(['Contract', 'PaymentMethod'])
      .size()
      .reset_index(name='CustomerCount')
      .sort_values('CustomerCount', ascending=False)
)

print("\nCROSS-SELLING OPPORTUNITIES")
print(cross_selling.head())
```

```python
total_customers = df['CustomerID'].nunique()
retained_customers = df[df['Churn'] == 0]['CustomerID'].nunique()

retention_rate = (retained_customers / total_customers) * 100

print("\nRETENTION RATE")
print(f"{retention_rate:.2f}%")
```

```python
import matplotlib.pyplot as plt
# Chart 1: TotalCharges by Tenure
# ===============================
tenure_revenue = df.groupby("Tenure")["TotalCharges"].sum().reset_index()

plt.figure()
plt.plot(tenure_revenue["Tenure"], tenure_revenue["TotalCharges"])
plt.xlabel("Tenure")
plt.ylabel("TotalCharges")
plt.title("TotalCharges Trend by Tenure")
plt.show()
```

```python
# Chart 2: Average MonthlyCharges by Tenure
# =====================================
avg_monthly = df.groupby("Tenure")["MonthlyCharges"].mean().reset_index()

plt.figure()
plt.plot(avg_monthly["Tenure"], avg_monthly["MonthlyCharges"])
plt.xlabel("Tenure")
plt.ylabel("MonthlyCharges")
plt.title("Average MonthlyCharges by Tenure")
plt.show()
```

```python
# Chart 3: Churn Rate by Tenure
# =====================================
churn_trend = df.groupby("Tenure")["Churn"].mean().reset_index()

plt.figure()
plt.plot(churn_trend["Tenure"], churn_trend["Churn"])
plt.xlabel("Tenure")
plt.ylabel("Churn Rate")
plt.title("Churn Rate by Tenure")
plt.show()
```

```python
# Chart 4: TotalCharges by Contract
# =====================================
contract_revenue = df.groupby("Contract")["TotalCharges"].sum().reset_index()

plt.figure()
plt.bar(contract_revenue["Contract"], contract_revenue["TotalCharges"])
plt.xlabel("Contract")
plt.ylabel("TotalCharges")
plt.title("TotalCharges by Contract")
plt.show()
```

```python
# Chart 5: TotalCharges by PaymentMethod
# =====================================
payment_revenue = df.groupby("PaymentMethod")["TotalCharges"].sum().reset_index()

plt.figure()
plt.bar(payment_revenue["PaymentMethod"], payment_revenue["TotalCharges"])
plt.xlabel("PaymentMethod")
plt.ylabel("TotalCharges")
plt.title("TotalCharges by Payment Method")
plt.xticks(rotation=30)
plt.show()
```
